```python
student = []

while True:

    print("\n===== Student Management System =====")
    print("1. Add Student")
    print("2. Display Student")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Delete Student")
    print("6. Total Students")
    print("7. Exit")

    choice = int(input("Enter the Choice: "))

    if choice == 1:

        name = input("Enter the Name: ")
        student.append(name)
        print("Student Added Successfully")

    elif choice == 2:

        if len(student) == 0:
            print("No Student Found")
        else:
            for i in student:
                print(i)

    elif choice == 3:

        name = input("Enter the Name: ")

        if name in student:
            print(name, "Student is Found")
        else:
            print("Student Not Found")

    elif choice == 4:

        old_name = input("Enter Old Student Name: ")

        if old_name in student:

            new_name = input("Enter New Student Name: ")

            index = student.index(old_name)
            student[index] = new_name

            print("Student Updated Successfully")

        else:
            print("Student Not Found")

    elif choice == 5:

        name = input("Enter the Name: ")

        if name in student:

            student.remove(name)
            print("Student Deleted Successfully")

        else:
            print("Student Not Found")

    elif choice == 6:

        print("Total Students =", len(student))

    elif choice == 7:

        print("Thank You!")
        break

    else:

        print("Invalid Choice")
```
