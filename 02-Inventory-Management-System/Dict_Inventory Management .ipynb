```python
inventory = {
    101: {
        "Name": "Laptop",
        "Category": "Electronics",
        "Brand": "HP",
        "Price": 65000,
        "Stock": 15,
        "Sold": 10
    },

    102: {
        "Name": "Mouse",
        "Category": "Electronics",
        "Brand": "Logitech",
        "Price": 800,
        "Stock": 40,
        "Sold": 25
    },

    103: {
        "Name": "Keyboard",
        "Category": "Electronics",
        "Brand": "Dell",
        "Price": 1500,
        "Stock": 20,
        "Sold": 12
    },

    104: {
        "Name": "Monitor",
        "Category": "Electronics",
        "Brand": "Samsung",
        "Price": 12000,
        "Stock": 8,
        "Sold": 6
    },

    105: {
        "Name": "Printer",
        "Category": "Office",
        "Brand": "Canon",
        "Price": 9000,
        "Stock": 5,
        "Sold": 4
    }
}

while True:

    print("\n" + "=" * 50)
    print("      INVENTORY MANAGEMENT SYSTEM")
    print("=" * 50)

    print("1. Display Products")
    print("2. Add Product")
    print("3. Search Product by ID")
    print("4. Search Product by Name")
    print("5. Update Price")
    print("6. Update Stock")
    print("7. Delete Product")
    print("8. Sell Product")
    print("9. Restock Product")
    print("10. Inventory Value")
    print("11. Revenue Report")
    print("12. Top Selling Product")
    print("13. Most Expensive Product")
    print("14. Cheapest Product")
    print("15. Low Stock Report")
    print("16. Out of Stock Report")
    print("17. Search By Price")
    print("18. Dashboard")
    print("19. Exit")

    choice = int(input("\nEnter Your Choice : "))

    # Display Products
   

    if choice == 1:

        if len(inventory) == 0:
            print("\nInventory is Empty.")

        else:

            print("\nPRODUCT LIST\n")

            for product_id, details in inventory.items():

                print("-" * 40)
                print("Product ID :", product_id)
                print("Name       :", details["Name"])
                print("Category   :", details["Category"])
                print("Brand      :", details["Brand"])
                print("Price      : ₹", details["Price"])
                print("Stock      :", details["Stock"])
                print("Sold       :", details["Sold"])
                print("-" * 40)

    # ======================================================
    # Add Product
    
    elif choice == 2:

        product_id = int(input("Enter Product ID : "))

        if product_id in inventory:

            print("Product ID already exists.")

        else:

            name = input("Enter Product Name : ")
            category = input("Enter Category : ")
            brand = input("Enter Brand : ")
            price = int(input("Enter Price : "))
            stock = int(input("Enter Stock : "))
            sold = int(input("Enter Sold Quantity : "))

            inventory[product_id] = {

                "Name": name,
                "Category": category,
                "Brand": brand,
                "Price": price,
                "Stock": stock,
                "Sold": sold

            }

            print("\nProduct Added Successfully.")

    # ======================================================
    # Search Product By ID
    # ======================================================

    elif choice == 3:

        product_id = int(input("Enter Product ID : "))

        if product_id in inventory:

            details = inventory[product_id]

            print("\nProduct Found\n")

            print("Name :", details["Name"])
            print("Category :", details["Category"])
            print("Brand :", details["Brand"])
            print("Price : ₹", details["Price"])
            print("Stock :", details["Stock"])
            print("Sold :", details["Sold"])

        else:

            print("Product Not Found.")

    # ======================================================
    # Search Product By Name
    # ======================================================

    elif choice == 4:

        product_name = input("Enter Product Name : ")

        found = False

        for product_id, details in inventory.items():

            if details["Name"].lower() == product_name.lower():

                print("\nProduct Found\n")

                print("Product ID :", product_id)
                print("Category :", details["Category"])
                print("Brand :", details["Brand"])
                print("Price : ₹", details["Price"])
                print("Stock :", details["Stock"])
                print("Sold :", details["Sold"])

                found = True
                break

        if not found:

            print("Product Not Found.")
```

```python
#Update Price
elif choice == 5:

    product_id = int(input("Enter Product ID : "))

    if product_id in inventory:

        new_price = int(input("Enter New Price : "))

        inventory[product_id]["Price"] = new_price

        print("Price Updated Successfully.")

    else:

        print("Product Not Found.")
#Update Stock
elif choice == 6:

    product_id = int(input("Enter Product ID : "))

    if product_id in inventory:

        new_stock = int(input("Enter New Stock : "))

        inventory[product_id]["Stock"] = new_stock

        print("Stock Updated Successfully.")

    else:

        print("Product Not Found.")
#Delete Product
elif choice == 7:

    product_id = int(input("Enter Product ID : "))

    if product_id in inventory:

        print("\nProduct Details")
        print("----------------------------")
        print("Name :", inventory[product_id]["Name"])
        print("Price :", inventory[product_id]["Price"])

        confirm = input("\nDelete this product? (yes/no) : ")

        if confirm.lower() == "yes":

            del inventory[product_id]

            print("Product Deleted Successfully.")

        else:

            print("Delete Cancelled.")

    else:

        print("Product Not Found.")
#Sell Product
elif choice == 8:

    product_id = int(input("Enter Product ID : "))

    if product_id in inventory:

        quantity = int(input("Enter Quantity To Sell : "))

        if quantity <= inventory[product_id]["Stock"]:

            inventory[product_id]["Stock"] -= quantity

            inventory[product_id]["Sold"] += quantity

            print("Product Sold Successfully.")

        else:

            print("Insufficient Stock.")

    else:

        print("Product Not Found.")
#Restock Product
elif choice == 9:

    product_id = int(input("Enter Product ID : "))

    if product_id in inventory:

        quantity = int(input("Enter Quantity To Add : "))

        inventory[product_id]["Stock"] += quantity

        print("Stock Updated Successfully.")

    else:

        print("Product Not Found.")
```

```python
# Inventory Value
elif choice == 10:

    total_value = 0

    for product in inventory.values():

        total_value += product["Price"] * product["Stock"]

    print("\nTotal Inventory Value : ₹", total_value)
#Revenue Report
elif choice == 11:

    total_revenue = 0

    for product in inventory.values():

        total_revenue += product["Price"] * product["Sold"]

    print("\nTotal Revenue : ₹", total_revenue)
# Top Selling Product
elif choice == 12:

    max_revenue = 0
    top_product = ""

    for product in inventory.values():

        revenue = product["Price"] * product["Sold"]

        if revenue > max_revenue:

            max_revenue = revenue
            top_product = product["Name"]

    print("\nTop Selling Product :", top_product)
    print("Revenue : ₹", max_revenue)
# Most Expensive Product
elif choice == 13:

    highest_price = 0
    expensive_product = ""

    for product in inventory.values():

        if product["Price"] > highest_price:

            highest_price = product["Price"]
            expensive_product = product["Name"]

    print("\nMost Expensive Product :", expensive_product)
    print("Price : ₹", highest_price)
✅ Cheapest Product
elif choice == 14:

    lowest_price = float("inf")
    cheapest_product = ""

    for product in inventory.values():

        if product["Price"] < lowest_price:

            lowest_price = product["Price"]
            cheapest_product = product["Name"]

    print("\nCheapest Product :", cheapest_product)
    print("Price : ₹", lowest_price)
# Low Stock Report
elif choice == 15:

    found = False

    print("\nLOW STOCK PRODUCTS\n")

    for product_id, product in inventory.items():

        if product["Stock"] < 10:

            print("----------------------------")
            print("Product ID :", product_id)
            print("Name :", product["Name"])
            print("Stock :", product["Stock"])

            found = True

    if not found:

        print("No Low Stock Products.")
# Out of Stock Report
elif choice == 16:

    found = False

    print("\nOUT OF STOCK PRODUCTS\n")

    for product_id, product in inventory.items():

        if product["Stock"] == 0:

            print("----------------------------")
            print("Product ID :", product_id)
            print("Name :", product["Name"])

            found = True

    if not found:

        print("No Product is Out of Stock.")
```
